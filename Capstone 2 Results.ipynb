```python
# this feels stupid, but, hey, practice:
# make a function to take an arbitrary dataframe, examine the contents of a column named 'assessment'
# it should contain dicts.  Read the dict, create a new column-value for each key-value pair in the dict

def expand_assessment(df):
    assessment = df.assessment
    new_df = pd.DataFrame(columns = assessment[0].keys())
    for assx in assessment:
        new_df = new_df.append(assx, ignore_index=True)
    return new_df
    
```
